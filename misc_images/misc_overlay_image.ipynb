```python
from PIL import Image
  
# Opening the primary image (used in background)
c9_team_averages_png = './output/c9/c9_team_averages.png'
img1 = Image.open(c9_team_averages_png)
  
# Opening the secondary image (overlay image)
img2 = Image.open('./misc_images/rank.png').convert("RGBA")
width, height = img2.size
img2 = img2.resize((width//3, height//3))

# Pasting img2 image on top of img1 
# starting at coordinates (0, 0)
img1.paste(img2, (540,15), mask = img2)
  
# Displaying the image
img1
```
